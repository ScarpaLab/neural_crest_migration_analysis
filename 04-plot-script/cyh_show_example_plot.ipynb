```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import sys
sys.path.append("..")
from utility import util
from utility import data_loader
from utility.plotter import CellPlotter
from utility.registration import DriftCorrector

metadata = util.load_metadata()
for k in range(len(metadata)):
    if k < 16:
        continue
    dl = data_loader.DataLoader(metadata.iloc[k])

    print(f'Processing subject {dl.sub}')
    drift = dl.load_data('drift')
    drift_corrector = DriftCorrector(dl.load_data('xyProjection')[:,1,:,:]) # Use Nuclei for drift calculation
    drift_corrector.shifts = drift

    nuc_label = dl.load_data('nucManualLabel')
    mem_label = dl.load_data('memManualLabel')
    dc_nuc_label = drift_corrector.apply_correction(nuc_label)
    dc_mem_label = drift_corrector.apply_correction(mem_label)
    dc_nuc_img = drift_corrector.apply_correction(dl.load_data('xyProjection')[:,1,:,:])
    dc_mem_img = drift_corrector.apply_correction(dl.load_data('xyProjection')[:,0,:,:])

    df_path = f'../03-extracted_params/cell_props_sub-{dl.sub:03d}.csv'
    kymo_path = f'../03-extracted_params/kymo_data_sub-{dl.sub:03d}.pkl'
    img_path = f'../05-plots/summary_images/sub-{dl.sub:03d}.svg'

    cells_df = pd.read_csv(df_path)
    kymo_data = util.load_pkl(kymo_path)

    cell_ids = np.array(list(kymo_data.keys()))
    n_cells = len(cell_ids)


    #plot_additive_composite(analyzer.kymo_data[6]['mem'], analyzer.kymo_data[6]['nuc'], ax=ax)
    cell_plotter = CellPlotter(
        cells_df, 
        kymo_data,
        dc_mem_img,
        dc_nuc_img
    )

    _, axes = plt.subplots(4,n_cells,figsize=(3*n_cells,10))

    for i, cell_idx in enumerate(cell_ids):

        cell_plotter.plot_movement_arrow(cell_idx=cell_idx, ax=axes[0,i], ch='mem')
        axes[0,i].set_title(f'Cell {cell_idx}')

        cell_plotter.plot_kymograph(cell_idx=cell_idx, ax=axes[1,i])

        cell_plotter.plot_cell_props(cell_idx=cell_idx, prop='pos_on_track', ax=axes[2,i], color = 'k', alpha=0.5)
        cell_plotter.plot_cell_props_scatter(cell_idx=cell_idx, prop='mem_front', ax=axes[2,i], marker='o', alpha=0.5)
        cell_plotter.plot_cell_props_scatter(cell_idx=cell_idx, prop='mem_back', ax=axes[2,i], marker='o', alpha=0.5)
        axes[2,i].legend()
        # Reverse the y axis
        axes[2,i].invert_yaxis()

        cell_plotter.plot_cell_props(cell_idx=cell_idx, prop='speed_nuc', ax=axes[3,i], alpha=0.8, color='k')
        axes[3,i].set_ylabel('Nuclear Speed', color='k')
        ax2 = axes[3,i].twinx()
        cell_plotter.plot_cell_props(cell_idx=cell_idx, prop='circularity_nuc', ax=ax2, alpha=0.8, color='r')
        ax2.set_ylabel('Nuclear Circularity', color='r')
        plt.tight_layout()

    sns.despine()
    plt.savefig(img_path, bbox_inches='tight')
    # Close the image to free up memory
    plt.close()
```
